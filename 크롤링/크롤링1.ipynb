```python
# 웹 페이지 요청을 위한 패키지
import requests

# 네이버 뉴스 페이지 URL
url = 'https://news.naver.com/'

# 네이버 뉴스 페이지 요청, 응답 저장
response = requests.get(url)
print(response.text)    # 내용 확인 -> 화면에서 보던 내용들은 cjs_t 뒤에 있음을 알 수 있음
```

```python
# 웹페이지 파싱을 위한 패키지   파싱: HTML에서 토큰? 구분별로 나누기 위해
from bs4 import BeautifulSoup

# response로부터 HTML 가져오기
html = response.text

# beautifulsoup으로 HTML 파싱
soup = BeautifulSoup(html, 'html.parser')

# HTML에서 원하는 태그 찾기
hdline_titles = soup.find_all('div', {'class':'cjs_t'})

# div에 cjs_t의 class에 속한 텍스트만 추출
for title in hdline_titles:
    print(title.get_text(strip=True))   # 결과 출력
```

#### 대한민국 정책 브리핑

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.korea.kr/news/top50List.do'
resp = requests.get(url)

soup = BeautifulSoup(resp.text)

top_news = []

# 크롤링할 태그 조건 설정
strong = soup.select('span.text > strong')  # text라는 클래스명을 가진 span 태그의 자식 중 strong 태그만 선택
for n in strong:
    top_news.append(n.get_text().strip())

print(top_news)
```

```python
from datetime import datetime

year = datetime.today().year    # 현재 연도 가져오기
month = datetime.today().month    # 현재 월 가져오기
day = datetime.today().day    # 현재 일 가져오기

print(f'>>>{year}년 {month}월 {day}일', end = ' ')
print('대한민국 정책 브리핑 TOP 50 뉴스>>>')

n = 1
for i in top_news[:50]:
    print('%d: '%n + i)
    n = n + 1
```
