#### 실제로 필요한 데이터 html에서 추출하는 작업

```python
# html 소스를 가져옵니다.
html = response.content
```

* html.parser: Python 표준 라이브러리에 포함된 파서입니다. 별도의 설치가 필요 없고 속도가 빠르며, 대다수의 경우에 충분히 잘 작동합니다. 하지만 복잡한 HTML 문서에서는 다른 파서들보다 덜 견고할 수 있습니다.
* lxml: 더 강력하고 빠른 파서입니다. HTML과 XML 문서를 모두 지원합니다. 사용하려면 lxml 라이브러리를 별도로 설치해야 합니다.
* html5lib: HTML5를 지원하는 파서로, HTML 문서를 웹 브라우저와 유사하게 처리합니다. 가장 느리지만, 가장 견고하게 HTML 문서를 처리합니다. html5lib 라이브러리도 별도로 설치해야 합니다.

```python
# BeautifulSoup을 사용하여 HTML을 파싱합니다. => 구조화된 데이터로 변환
soup = BeautifulSoup(html, 'html.parser')
```

```python
articles = soup.find_all('a', class_ = 'sa_text_title')
```

find_all(1, 2)
* 1번 인자: 찾고자 하는 태그 이름입니다. 여기서는 'a' 태그를 찾고 있습니다. a 태그는 일반적으로 하이퍼링크
* 2번 인자: 속성(attribute)을 조건으로 설정할 수 있습니다. 딕셔너리 형태로 제공하며, 여기서는 'class': 'nclicks(fls.page)'라는 조건을 사용

```python
# 찾은 기사들의 제목과 링크를 출력
for article in articles:
    # <strong> 태그 내의 텍스트 가져오기
    title = article.find('strong', class_='sa_text_strong').text.strip()
    # <a> 태그의 href 속성 가져오기
    link = article['href']
    print(f"제목: {title}\n링크: {link}\n")
```

#### Selenium
웹 브라우저를 자동으로 제어할 수 있는 도구
1. 브라우저 자동화: Selenium은 실제 브라우저(예: Chrome, Firefox 등)를 열어 웹 페이지를 탐색하고 조작할 수 있습니다.
2. 다양한 언어 지원: Python, Java, C#, Ruby 등 여러 프로그래밍 언어를 지원합니다.
3. 다양한 브라우저 지원: Chrome, Firefox, Safari, Edge 등 여러 브라우저에서 동작합니다.
4. 웹 애플리케이션 테스트: Selenium은 웹 애플리케이션을 자동으로 테스트하는 데 자주 사용됩니다. 다양한 시나리오를 설정하고 테스트할 수 있습니다.

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

# selenium으로 키를 조작하기 위한 import
from selenium.webdriver.common.keys import Keys

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 페이지 로딩을 기다리는데에 사용할 time 모듈 import
import time
```

```python
# Chrome 브라우저 열기
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # 브라우저 창을 띄우지 않음
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
# 크롬드라이버 실행
# driver = webdriver.Chrome(options = options) 
driver = webdriver.Chrome()

#크롬 드라이버에 url 주소 넣고 실행
driver.get('https://www.google.co.kr/')

# 페이지가 완전히 로딩되도록 3초동안 기다림
time.sleep(3)
```

```python
# 네이버 뉴스 특정 날짜의 URL
url = 'https://news.naver.com/breakingnews/section/101/263?date=20240709'
driver.get(url)
```

```python
# '기사 더보기' 버튼을 클릭하여 모든 기사 로드
while True:
    try:
        # '기사 더보기' 버튼이 로드될 때까지 기다림
        more_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "_CONTENT_LIST_LOAD_MORE_BUTTON"))
        )
        more_button.click()
        time.sleep(2)  # 기사를 로드하는 시간을 고려하여 잠시 대기
    except Exception as e:
        print("더 이상 '기사 더보기' 버튼이 없습니다.", str(e))
        break
```

```python
# 페이지 소스를 가져와서 BeautifulSoup으로 파싱
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
# class가 'sa_text_title'인 모든 <a> 태그 찾기
articles = soup.find_all('a', class_='sa_text_title')
```

```python
# 찾은 기사들의 제목과 링크를 출력
data = []
for article in articles:
    title = article.find('strong', class_='sa_text_strong').text.strip()
    link = article['href']
    print(f"제목: {title}\n링크: {link}\n")
    data.append({'Title': title, 'Link': link})
```

```python
# 데이터프레임 생성
df = pd.DataFrame(columns=['Title', 'Link'])
```

```python
# 리스트를 데이터프레임으로 변환하여 기존 데이터프레임과 결합
df = pd.concat([df, pd.DataFrame(data)], ignore_index=True)
```

```python
def ex_tag(sid, page):
    ### 뉴스 분야(sid)와 페이지(page)를 입력하면 그에 대한 링크들을 리스트로 추출하는 함수 ###
    
    ## 1.
    url = f"https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1={sid}"\
    "#&date=%2000:00:00&page={page}"
    html = requests.get(url, headers={"User-Agent": "Mozilla/5.0"\
    "(Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) "\
    "Chrome/110.0.0.0 Safari/537.36"})
    soup = BeautifulSoup(html.text, "lxml")
    a_tag = soup.find_all("a")
    
    ## 2.
    tag_lst = []
    for a in a_tag:
        if "href" in a.attrs:  # href가 있는것만 고르는 것
            if (f"sid={sid}" in a["href"]) and ("article" in a["href"]):
                tag_lst.append(a["href"])
                
    return tag_lst
```
