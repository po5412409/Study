## 1. 시계열데이터

```python
import pandas as pd
import numpy as np
```

```python
# datetime index

dates = pd.date_range('2020-01-01', periods = 48, freq='M')
dates
```

```python
timestamp = np.arange(len(dates))
timestamp
```

```python
trend_factor = timestamp*1.1
trend_factor
```

```python
cycle_factor = 10 * np.sin(np.linspace(0, 3.14*2, 48))
cycle_factor
```

```python
seasonal_factor = 7 * np.sin(np.linspace(0, 3.14*8, 48))
seasonal_factor
```

```python
np.random.seed(2004)
irregular_factor = 2 * np.random.randn(len(dates))
irregular_factor
```

```python
df = pd.DataFrame({'timeseries' : trend_factor + cycle_factor + seasonal_factor + irregular_factor,
                'trend' : trend_factor,
                'cycle' : cycle_factor,
                'seasonal' : seasonal_factor,
                'irregular_factor' : irregular_factor},
                index = dates)
df
```

## 2. 그래프 시각화

### 시계열 데이터

```python
import matplotlib.pyplot as plt

plt.figure(figsize=[10,6])
df.timeseries.plot()
plt.title('Timeseries (Additive Model)', fontsize = 16)
plt.ylim(-12, 55)
plt.show()
```

### 추세변동

```python
plt.figure(figsize = [10, 6])
df.trend.plot()
plt.title('Trend Factor', fontsize = 16)
plt.ylim(-12, 55)
plt.show()
```

추세선은 시간의 흐름에 따라 직선의 상승선을 보임

### 순환 변동

```python
plt.figure(figsize = [10, 6])
df.cycle.plot()
plt.title('Cycle Factor', fontsize = 16)
plt.ylim(-12, 55)
plt.show()
```

### 계절 변동

```python
plt.figure(figsize = [10, 6])
df.seasonal.plot()
plt.title('Seasonal Factor', fontsize = 16)
plt.ylim(-12, 55)
plt.show()
```
